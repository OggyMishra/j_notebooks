```python
import pandas as pd
df = pd.read_csv('Noida2.csv', encoding='unicode_escape')
new_header = df.iloc[0]
df = df[1:]
df.columns = new_header 
columns = list(df.columns)
columns = [x for x in columns if str(x) != 'nan' and str(x) != 'Correct Answers']
columns
for col in columns:
    df[col] = df[col] == col
df.head() 
df['Correct Answer'] = (df == True).sum(axis =1)
df.to_csv('result.csv')
```
