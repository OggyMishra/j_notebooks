# Perform merge operations

```python
import pandas as pd
# create two dataframe first
employee = ['John','Andrew','Roy','Sam']
salary = [25000, 35000, 23400, 43000]
dept = ['dept1', 'dept2', 'dept3','dept4']
df = pd.DataFrame({'employee': employee, 'salary': salary, 'dept': dept})
df
```

```python
deptTbArray = np.array([['dept1', 'dept2','dept5', 'dept6', 'dept7'], ['HR','IT', 'Admin', 'Eng', 'Content']])
deptTbArray
```

```python
df2 = pd.DataFrame({'dept': deptTbArray[0], 'deptName': deptTbArray[1]})
df2
```

```python
# inner merge
df2.merge(df, on='dept', how='inner')
```

```python
# outer merge
df2.merge(df, on = 'dept', how = 'outer')
```

```python
import pandas as pd
user_usage = pd.read_csv('user_usage.csv')
```

```python
user_usage.head()
```

```python
user_device = pd.read_csv('user_device.csv')
```

```python
user_device.head()
```

```python
android_devices = pd.read_csv('android_devices.csv')
android_devices.head()
```

```python
device_usage = user_device.merge(user_usage[['outgoing_mins_per_month','outgoing_sms_per_month','use_id']], how = 'inner', on = 'use_id')
device_usage.head()
```

```python
android_devices.rename(columns={'Retail Branding': 'manufacturer'})
final_res = device_usage.merge(android_devices, how='outer', left_on='device', right_on='Model', indicator=True)
print()
final_res
```

```python
res = final_res[['use_id', 'user_id', 'platform','device','outgoing_mins_per_month', 'outgoing_sms_per_month', 'Retail Branding']]
```

```python
res.groupby('Retail Branding').agg({
    'outgoing_mins_per_month': 'min',
    'outgoing_sms_per_month': 'mean',
    'use_id': 'count'
})
```
